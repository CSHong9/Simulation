This following code implements Thomas C. Schelling. 1978. “Thermostats, Lemons, and Other Families of Models.” Micromotives and Macrobehavior. (pgs 67-69)

Replication of Ski Lift Model by Chen-Shuo Hong (2017) <br />

This model is an illustration of conservative system (semi-closed system). Namely, there is a system "that gains or loses the things we are keeping track of at a limited number of entry and exist points at which, if we wish, we can count them" (p.69)

Pseudocode in plain English
1. Create three lists (waiting line, ski lift, and the skiing phase)
2. People initially wait in a line.
3. Every six second, some people are taking the lift and the same number of people are coming down.
3. Change the speed of lift and watch how many people stay in the waiting zone.

Note: this replication is system-level, deterministic, synchronous, continuous-state implementation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set Parameters
size = 100 # populaion size
speed_max = 30 # the maximum speed of lift
length = 120 # the length of ski lift
X3 = 50 # a constant of people in skiing phase
```

```python
# Set Initial Conditions
#X1 = 0 # initial number of people in wailing line
#X2 = 0 # initial number of people in ski lift
```

```python
# BEGIN Loop Over Iterations
def skilift(size, length, speed_max):    
    k = np.arange(10, speed_max + 1)
    maxt = len(k) + 1
    X1 = np.zeros(maxt)
    X2 = np.zeros(maxt)
    X1[0] = size
    X2[0] = 0
    for i in range(0, maxt - 1):
        X2[i + 1] = (2 * length) / k[i]
        X1[i + 1] = size - (X2[i + 1] + X3)
#    print(k, X1, X2)
    plt.plot(k, X2[1:maxt], 'b', label = "Number of people taking the ski lift")
    plt.plot(k, X1[1:maxt], 'r', label = "Number of people in waiting line")
```

```python
skilift(size, length, speed_max)
plt.xlabel('Speed')
plt.ylabel('Number of people in waiting line')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # plot legend outside of graph
plt.show()
```
